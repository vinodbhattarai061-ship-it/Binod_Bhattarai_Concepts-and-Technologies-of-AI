```python
# 4 (Task 1)
import numpy as np

''' Different Types of Array '''

# 1. Initialize an empty array with size 2x2
empty_array = np.empty((2, 2))
print("1. Empty Array 2x2:\n", empty_array)

# 2. Initialize an all-one array with size 4x2
ones_array = np.ones((4, 2))
print("\n2. Ones Array 4x2:\n", ones_array)

# 3. Return new array with given shape and type, filled with a value
filled_array = np.full((3, 3), 7, dtype=int)
print("\n3. Filled Array 3x3 with 7:\n", filled_array)

# 4. Return a new array of zeros with same shape and type as a given array
zeros_like_array = np.zeros_like(filled_array)
print("\n4. Zeros Like Filled Array:\n", zeros_like_array)

# 5. Return a new array of ones with same shape and type as a given array
ones_like_array = np.ones_like(filled_array)
print("\n5. Ones Like Filled Array:\n", ones_like_array)

# 6. Convert list to numpy array
new_list = [1, 2, 3, 4]
array_from_list = np.array(new_list)
print("\n6. Numpy array from list:\n", array_from_list)
```

```python
# 4 (Task 2)

''' Manipulation in Array '''

# 1. Array with values 10 to 49
arr_10_49 = np.arange(10, 50)
print("\n1. Array 10 to 49:\n", arr_10_49)

# 2. 3x3 matrix with values 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n2. 3x3 Matrix 0-8:\n", matrix_3x3)

# 3. 3x3 identity matrix
identity_3x3 = np.eye(3)
print("\n3. Identity Matrix 3x3:\n", identity_3x3)

# 4. Random array of size 30 and mean
rand_array = np.random.random(30)
print("\n4. Random Array of size 30:", rand_array)
print("Mean:", rand_array.mean())

# 5. 10x10 random array and min/max
rand_10x10 = np.random.random((10, 10))
print("\n5. 10x10 Random Array Min:", rand_10x10.min(), "Max:", rand_10x10.max())

# 6. Zero array of size 10, replace 5th element with 1
zero_arr = np.zeros(10)
zero_arr[4] = 1
print("\n6. Zero array with 5th element as 1:", zero_arr)

# 7. Reverse array
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\n7. Reversed Array:", reversed_arr)

# 8. 2D array with 1 on border and 0 inside
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n8. Border Array:\n", border_array)

# 9. 8x8 checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\n9. 8x8 Checkerboard Pattern:\n", checkerboard)
```

```python
# 4 (Task 3)

''' Operations in Array '''

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add x + y
print("\n1. x + y:\n", x + y)

# 2. Subtract x - y
print("\n2. x - y:\n", x - y)

# 3. Multiply array with integer (example *3)
print("\n3. x * 3:\n", x * 3)

# 4. Square of each element
print("\n4. x squared:\n", x ** 2)

# 5. Dot products
print("\n5a. v.w =", np.dot(v, w))
print("5b. x.v =", np.dot(x, v))
print("5c. x.y =\n", np.dot(x, y))

# 6. Concatenate along rows and columns
print("\n6a. Concatenate x and y along rows:\n", np.concatenate((x, y), axis=0))
print("6b. Concatenate v and w along column:\n", np.column_stack((v, w)))


# 7. Concatenate x and v (will give error)
try:
    np.concatenate((x, v), axis=0)
except ValueError as e:
    print("\n7. Concatenating x and v Error:", e)
    print("Reason: x is 2D and v is 1D, shapes not compatible for concatenation along axis 0")
```

```python
# 4 (Task 4)

''' Matrix Operation using Array '''

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A.A^-1 = I
A_inv = np.linalg.inv(A)
print("\n1. A * A_inv =\n", np.dot(A, A_inv))

# 2. Prove AB != BA
print("\n2. AB =\n", np.dot(A, B))
print("   BA =\n", np.dot(B, A))
print("AB != BA:", not np.allclose(np.dot(A, B), np.dot(B, A)))

# 3. Prove (AB)^T = B^T A^T
AB_T = np.dot(A, B).T
B_T_A_T = np.dot(B.T, A.T)
print("\n3. (AB)^T =\n", AB_T)
print("   B^T A^T =\n", B_T_A_T)
print("Equality:", np.allclose(AB_T, B_T_A_T))

# Solve linear equations using matrix inverse
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

coeff_matrix = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
constants = np.array([-1, -3, 9])

# Solve using np.linalg.inv
solution = np.dot(np.linalg.inv(coeff_matrix), constants)
print("\nSolution using inverse:\n", solution)
```
