```python
# Daily time (in hours): [study, entertainment, sleep]
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

#  7 (Task 1)

low_study = []        # < 3 hours
moderate_study = []   # 3 to 5 hours
high_study = []       # > 5 hours

for study, entertainment, sleep in time_data:
    if study < 3:
        low_study.append(study)
    elif 3 <= study <= 5:
        moderate_study.append(study)
    else:
        high_study.append(study)
print("\nTask 1 Results:")
print()
print("Low Study Hours (<3):", low_study)
print("Moderate Study Hours (3–5):", moderate_study)
print("High Study Hours (>5):", high_study)

# 7 (Task 2)

print("\nTask 2 Results:")
print()
print("Days with Low Study Time:", len(low_study))
print("Days with Moderate Study Time:", len(moderate_study))
print("Days with High Study Time:", len(high_study))

# 7 (Task 3)

study_minutes = []

for study, entertainment, sleep in time_data:
    study_minutes.append(study * 60)
print("\nTask 3 Result:")
print()
print("\nStudy Hours Converted to Minutes:")
print(study_minutes)
```

```python
# 7 (Task 4)

''' Creating Three Different Types of Array '''

study_hours = []
entertainment_hours = []
sleep_hours = []

''' Using Loop to Append data in Array '''

for study, entertainment, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(entertainment)
    sleep_hours.append(sleep)

''' Calculating The Sum and Then Dividing it By The Length '''

avg_study = sum(study_hours) / len(study_hours)
avg_entertainment = sum(entertainment_hours) / len(entertainment_hours)
avg_sleep = sum(sleep_hours) / len(sleep_hours)

''' Printing Out The Results '''

print("\nTask 4 Results:")
print("\nAverage Time Use:")
print()
print("Average Study Hours:", round(avg_study, 2))
print("Average Entertainment Hours:", round(avg_entertainment, 2))
print("Average Sleep Hours:", round(avg_sleep, 2))
```

```python
# 8 1 (Task 1)
def sum_nested_list(nested_list):
    ''' Recursively sum all numbers in a nested list. '''

    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)  # Recursive call
        else:
            total += element
    return total


# Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("\nTask 1 Result:")
print()
print("Output:", sum_nested_list(nested_list))
```

```python
# 8 1 (Task 2)
''' Function Used For Generating Permutations of a String. '''

def generate_permutations(s):

    '''Generating  all unique permutations of a string using recursion.'''

    if len(s) == 1:
        return [s]

    permutations = []
    for i in range(len(s)):
        current_char = s[i]
        remaining = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining):
            new_perm = current_char + perm
            if new_perm not in permutations:  # avoid duplicates (for "aab")
                permutations.append(new_perm)

    return permutations


# Output
print("\nTask 2 Results:")
print()
print("Output (Without Duplicates Handling):", generate_permutations("abc"))
print("Output (With Duplicates Handled):", generate_permutations("aab"))
```

```python
#8 1 (Task 3)

''' To Know The Size of Directory '''

def calculate_directory_size(directory):
    total_size = 0

    for name, item in directory.items():
        if isinstance(item, dict):
            # If it's a subdirectory → recursively calculate its size
            total_size += calculate_directory_size(item)
        else:
            # It's a file → add its size
            total_size += item

    return total_size
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}
print("\nTask3 Result:")
print("Total directory size:", calculate_directory_size(directory_structure), "KB")
```

```python
# 8 2 (Task 1)
def min_coins(coins, amount):

    ''' Uses dynamic programming '''

    # Initialize DP array
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # base case: 0 coins needed for amount 0

    # Build the table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1


# Testing

coins = [1, 2, 5]
amount = 11
print("\nTask1 Result:")
print("Output:", min_coins(coins, amount))  # Expected: 3
```

```python
def longest_common_subsequence(s1, s2):

    ''' Returns the length of the Longest Common Subsequence (LCS) '''

    m, n = len(s1), len(s2)
    # DP table initialization
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Build table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]


# Testing

s1 = "abcde"
s2 = "ace"
print("\nTask 2 Result:")
print("Output:", longest_common_subsequence(s1, s2))  # Expected: 3
```

```python
def knapsack(weights, values, capacity):

    ''' Uses Dynamic Programming Concept '''

    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Build table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]


# Test

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print("\nTask3 Result:")
print("Output:", knapsack(weights, values, capacity))  # Expected: 9
```
