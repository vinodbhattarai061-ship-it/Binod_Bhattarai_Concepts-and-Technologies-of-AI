```python
# Problem - 2: Subsetting Rows
''' 1. Load dataset '''
df = pd.read_csv("Titanic-Dataset.csv")
print("First Output:\n")
print("Dataset Loaded Successfully")
print(df.head())

''' 2. Subsetting Rows '''

# (a) Fare greater than 100
fare_gt_100 = df[df['Fare'] > 100]
print("\nSecond(a) Output: Passengers with Fare > 100")
print(fare_gt_100)

# (b) Passengers in first class (Pclass = 1)
first_class = df[df['Pclass'] == 1]
print("\nSecond(b) Output: First Class Passengers (Pclass = 1)")
print(first_class)

# (c) Female passengers under 18
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print("\nSecond(c) Output: Female Passengers under 18")
print(female_under_18)

''' 3. Subsetting Rows by Categorical Variables '''

# (d) Passengers embarked at "C" or "S"
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print("\nThird(d) Output: Passengers Embarked at C or S")
print(embarked_c_or_s)

# (e) Passengers in first or second class (Pclass in [1, 2])
first_second_class = df[df['Pclass'].isin([1, 2])]
print("\nThird(e) Output: Passengers in First or Second Class (Pclass 1 or 2)")
print(first_second_class)
```

```python
# Task (3.2) - EDA Practice Exercise 1

''' 1. Load dataset '''
df = pd.read_csv("Titanic-Dataset.csv")
print("First Output:\n")
print("Dataset Loaded Successfully")
print(df.head())

''' 2. Handle missing Age values '''
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
print("\nSecond Output: Missing Age values filled with median")
print(df['Age'].head())

''' 3. Add Fare per Year column (Fare / Age) '''
df['fare_per_year'] = df['Fare'] / df['Age']
print("\nThird Output: Added fare_per_year Column")
print(df[['Name', 'Age', 'Fare', 'fare_per_year']].head())

''' 4. Subset rows where fare_per_year > 5 '''
high_fare_age = df[df['fare_per_year'] > 5]
print("\nFourth Output: Passengers with fare_per_year > 5")
print(high_fare_age)

''' 5. Sort high_fare_age by descending fare_per_year '''
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
print("\nFifth Output: Sorted high_fare_age by descending fare_per_year")
print(high_fare_age_srt)

''' 6. Select only Name and fare_per_year columns '''
result = high_fare_age_srt[['Name', 'fare_per_year']]
print("\nSixth Output: Passenger Name and fare_per_year")
print(result)
```

```python
# Task (3.2) - EDA Practice Exercise 2
import matplotlib.pyplot as plt

''' 1. Load dataset '''
df = pd.read_csv("Titanic-Dataset.csv")
print("First Output:\n")
print("Dataset Loaded Successfully")
print(df.head())

''' 2. Handle missing Age values '''
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
print("\nSecond Output: Missing Age values filled with median")
print(df['Age'].head())

''' 3. Add fare_per_class column (Fare / Pclass) '''
df['fare_per_class'] = df['Fare'] / df['Pclass']
print("\nThird Output: Added fare_per_class Column")
print(df[['Name', 'Pclass', 'Fare', 'fare_per_class']].head())

''' 4. Subset adult male passengers (Age >= 18 and Sex == 'male') '''
adult_males = df[(df['Age'] >= 18) & (df['Sex'] == 'male')]
print("\nFourth Output: Adult Male Passengers")
print(adult_males)

''' 5. Sort adult males by descending fare_per_class '''
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)
print("\nFifth Output: Sorted Adult Males by fare_per_class")
print(adult_males_srt)

''' 6. Select Name, Age, fare_per_class columns '''
result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
print("\nSixth Output: Name, Age, and fare_per_class of Adult Males")
print(result)

''' 7. Plot top 10 adult males paying highest fare relative to class '''
top_10 = result.head(10)
plt.figure(figsize=(10,6))
plt.barh(top_10['Name'], top_10['fare_per_class'], color='skyblue')
plt.xlabel('Fare per Class (Fare / Pclass)')
plt.ylabel('Passenger Name')
plt.title('Top 10 Adult Male Passengers by Fare per Class')
plt.gca().invert_yaxis()  # Highest fare on top
plt.tight_layout()
plt.show()
```

```python
# Task (3.3) - EDA with Group-by Method and Graph
''' 1. Load dataset '''
df = pd.read_csv("Titanic-Dataset.csv")
print("First Output:\n")
print("Dataset Loaded Successfully")
print(df.head())

''' 2. Handle missing Age values if needed '''
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

''' 3. Calculate total Fare across all passengers '''
total_fare = df['Fare'].sum()
print("\nSecond Output: Total Fare across all passengers")
print(total_fare)

''' 4. Calculate total Fare by passenger class '''
fare_first_class = df[df['Pclass'] == 1]['Fare'].sum()
fare_second_class = df[df['Pclass'] == 2]['Fare'].sum()
fare_third_class = df[df['Pclass'] == 3]['Fare'].sum()
print("\nThird Output: Total Fare by Class")
print(f"First Class: {fare_first_class}, Second Class: {fare_second_class}, Third Class: {fare_third_class}")

''' 5. Combine totals into a list '''
fare_totals = [fare_first_class, fare_second_class, fare_third_class]

''' 6. Calculate percentage of total fare revenue by class '''
fare_percentages = [(fare / total_fare) * 100 for fare in fare_totals]
print("\nFourth Output: Percentage of Total Fare Revenue by Class")
print(fare_percentages)

''' 7. Create a summary DataFrame '''
fare_summary = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'Total_Fare': fare_totals,
    'Percentage_of_Total_Fare': fare_percentages
})
print("\nFare Revenue Summary by Class:")
print(fare_summary)

''' 8. Plot a bar chart of fare percentage by class '''
plt.figure(figsize=(8,5))
plt.bar(fare_summary['Pclass'].astype(str), fare_summary['Percentage_of_Total_Fare'], color=['gold','silver','brown'])
plt.xlabel('Passenger Class')
plt.ylabel('Percentage of Total Fare Revenue')
plt.title('Percentage of Total Fare Revenue by Passenger Class')
plt.ylim(0, max(fare_percentages)+10)
for i, v in enumerate(fare_percentages):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center')
plt.show()
```

```python
# Task (3.4) - Passenger Age Group Analysis
''' 1. Load dataset '''
df = pd.read_csv("Titanic-Dataset.csv")
print("First Output:\n")
print("Dataset Loaded Successfully")
print(df.head())

''' 2. Handle missing Age values '''
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

''' 3. Create a new column 'age_group' '''
def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(categorize_age)
print("\nSecond Output: Added age_group Column")
print(df[['Name', 'Age', 'age_group']].head())

''' 4. Calculate total number of passengers '''
total_passengers = len(df)
print("\nThird Output: Total Number of Passengers")
print(total_passengers)

''' 5. Count passengers in each age group '''
age_group_counts = df['age_group'].value_counts()
print("\nFourth Output: Count of Passengers in Each Age Group")
print(age_group_counts)

''' 6. Calculate proportion and percentage '''
age_group_percentage = (age_group_counts / total_passengers) * 100
print("\nFifth Output: Percentage of Passengers in Each Age Group")
print(age_group_percentage)

''' 7. Plot the percentage of passengers by age group '''
plt.figure(figsize=(6,5))
plt.bar(age_group_percentage.index, age_group_percentage.values, color=['skyblue','orange','green'])
plt.xlabel('Age Group')
plt.ylabel('Percentage of Passengers')
plt.title('Percentage of Passengers by Age Group on Titanic')
for i, v in enumerate(age_group_percentage.values):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center')
plt.ylim(0, max(age_group_percentage.values)+10)
plt.show()
```
